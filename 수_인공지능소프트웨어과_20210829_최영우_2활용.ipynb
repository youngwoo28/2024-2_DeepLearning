```python
# library call
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics

names = ['sepal-length', 'sepal-width','petal-lengh', 'petal-width', 'species']

dataset = pd.read_csv('../content/dmu_data/Iris.csv', names=names)
```

```python
dataset
```

```python
# training & testdataset
X = dataset.iloc[1:, :-1].values   # 모든 행은 사용하지만, 열(칼럼)은 뒤에서 하나를 뺸 값을 가져와서 X에 저장
y = dataset.iloc[1:, 4].values     # 모든 행은 사용하지만, 열은 앞에서 다섯 번쨰 값만 가져와서 y에 저장

# training & testdataset ratio
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)   # testdataset 20%

from sklearn.preprocessing import StandardScaler
s = StandardScaler()
s.fit(X_train)

X_train = s.transform(X_train)
X_test = s.transform(X_test)
```

```python
# create model : KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train, y_train)
```

```python
# model accuary
from sklearn.metrics import accuracy_score
y_pred = knn.predict(X_test)
print("정확도: {}".format(accuracy_score(y_test, y_pred)))
```

```python
########################################################
```

```python
k = 50
acc_array = np.zeros(k)
for k in np.arange(5, k+1, 2):
    classifier = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
    y_pred = classifier.predict(X_test)
    acc = metrics.accuracy_score(y_test, y_pred)
    acc_array[k-1] = acc

max_acc = np.amax(acc_array)
acc_list = list(acc_array)
k = acc_list.index(max_acc)
print("정확도", max_acc, "으로 최적의 k는", k+1, "입니다.")
```
